```python
import os
import cv2


DATASET = "DATASET"


for folder in os.listdir(DATASET):
    folderpath = os.path.join(DATASET, folder)
    
    
    if os.path.isdir(folderpath):
        print(f"Cleaning directory: {folder}")
        
        
        for posture in os.listdir(folderpath):
            posture_path = os.path.join(folderpath, posture)
            
            
            if os.path.isdir(posture_path):
                print(f"Processing posture folder: {posture}")
                
                
                for image in os.listdir(posture_path):
                    img_path = os.path.join(posture_path, image)
                    
                    # Check if it's an actual file
                    if os.path.isfile(img_path):
                        try:
                            
                            # print(f"Reading image: {img_path}")
                            
                            # Read the image
                            img = cv2.imread(img_path)
                            
                            # If the image is not valid (corrupt), remove it
                            if img is None:
                                print(f"Removing Corrupt Image: {img_path}")
                                os.remove(img_path)
                            else:
                                print("Clean1 done")
                        except Exception as e:
                            print(f"Error reading image {img_path}: {e}")
                            os.remove(img_path)
```

```python
# Resize the image
import os
import cv2


DATASET = "DATASET"
Resized_Folder = "resized_images"


for folder in os.listdir(DATASET):
    folderpath = os.path.join(DATASET, folder)
    
    if os.path.isdir(folderpath):
        
        for posture in os.listdir(folderpath):
            posture_path = os.path.join(folderpath, posture)
            
            if os.path.isdir(posture_path):
                print(f"Processing posture folder: {posture}")
                
                
                resized_posture_path = os.path.join(Resized_Folder, folder, posture)
                os.makedirs(resized_posture_path, exist_ok=True)
                
                
                for image in os.listdir(posture_path):
                    img_path = os.path.join(posture_path, image)
                    
                    if os.path.isfile(img_path):  # Ensure it's a file
                        try:
                            
                            img = cv2.imread(img_path)
                            
                            
                            if img is None:
                                print(f"Skipping corrupt image: {img_path}")
                                continue
                            
                            
                            resizeimage = cv2.resize(img, (224, 224))
                            
                            
                            newpath = os.path.join(resized_posture_path, image)
                            cv2.imwrite(newpath, resizeimage)
                            print(f"Resized and saved image: {newpath}")
                            
                        except Exception as e:
                            print(f"Error processing image {img_path}: {e}")
```

```python
#label encoding

DATASET = "resized_images"  
label_mapping = {  
    "downdog": 0,
    "goddess": 1,
    "plank": 2,
    "tree": 3,
    "warrior2": 4,
}


data = []


for pose_name, label in label_mapping.items():
    pose_folder = os.path.join(DATASET, pose_name)
    
    if os.path.isdir(pose_folder):  
        print(f"Processing folder: {pose_name}")
        for image in os.listdir(pose_folder):
            img_path = os.path.join(pose_folder, image)
            if os.path.isfile(img_path):
                # Append the image path and label to the data list
                data.append([img_path, label])


data_array = np.array(data)


np_file = "labeled_data.npy"
np.save(np_file, data_array)

print(f"Label encoding completed. Saved to {np_file}.")
```
